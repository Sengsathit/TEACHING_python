## 🧱 1. Fonctions personnalisées

Les fonctions permettent d’organiser votre code en blocs réutilisables.

```python
def say_hello(nom):
    print(f"Bonjour {nom} !")

say_hello("Alice")
```

```python
# Fonction avec valeur de retour
def addition(a, b):
    return a + b

resultat = addition(3, 4)
print(resultat)
```

___

## 🧩 2. Listes

Les listes sont des collections ordonnées et modifiables d’objets.

```python
fruits = ["pomme", "banane", "cerise"]
print(fruits)
print(fruits[1])  # accès par index

fruits.append("orange")
print(fruits)

for fruit in fruits:
    print(fruit)
```

___

## 🧩 3. Dictionnaires

Les dictionnaires stockent des paires clé/valeur.

```python
contact = {"nom": "Alice", "email": "alice@example.com"}
print(contact["nom"])
contact["téléphone"] = "0123456789"

for cle, valeur in contact.items():
    print(cle, ":", valeur)
```

___

## 📂 4. Lire et écrire dans un fichier texte

Nous pouvons stocker et récupérer des informations dans des fichiers avec les fonctions `open`, `read`, `write`, et le mot-clé `with`.

```python
# Écriture dans un fichier
with open("exemple.txt", "w") as f:
    f.write("Bonjour, ceci est un fichier texte.\n")
    f.write("On peut y écrire plusieurs lignes.")
```

```python
# Lecture du fichier
with open("exemple.txt", "r") as f:
    contenu = f.read()
    print(contenu)
```

___

## 🛠️ 5. Manipulation de données

Utiliser les fonctions et structures vues pour traiter des informations simples.

```python
# Exemple : compter les mots d'une phrase
phrase = "Python est un langage de programmation formidable"
mots = phrase.split()
print("Nombre de mots:", len(mots))
print("Liste des mots:", mots)
```
