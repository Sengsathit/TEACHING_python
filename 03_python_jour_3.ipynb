# 📗 Python - Jour 3 : Programmation Orientée Objet (POO)

Bienvenue dans cette troisième journée. Aujourd’hui, nous allons découvrir la Programmation Orientée Objet (POO), un paradigme essentiel dans le développement logiciel.

🔍 **Objectif** : Être capable de créer des classes, des objets, des méthodes et de structurer un programme avec ces concepts.

___

## 🧭 Pourquoi la POO ?

La POO permet :
- de **modéliser** des objets du monde réel,
- de **structurer** le code de manière plus claire et maintenable,
- de **réutiliser** du code grâce à l’héritage.

Elle repose sur 4 piliers :
1. **Encapsulation**
2. **Abstraction**
3. **Héritage**
4. **Polymorphisme**

___

## 🧱 1. Créer une classe et instancier des objets

```python
class Person:
    def __init__(self, nom, age):
        self.nom = nom
        self.age = age

    def introduce_myself(self):
        print(f"Bonjour, je m'appelle {self.nom} et j'ai {self.age} ans.")

# Création d'une instance
alice = Person("Alice", 30)
alice.introduce_myself()
```

___

## 🔧 2. Ajouter des comportements (méthodes)

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.speed = 0

    def accelerate(self):
        self.speed += 10
        print(f"La voiture roule à {self.speed} km/h.")

    def slow_down(self):
        self.speed = max(0, self.speed - 10)
        print(f"Vitesse réduite à {self.speed} km/h.")

my_car = Car("Toyota", "Corolla")
my_car.accelerate()
my_car.accelerate()
my_car.slow_down()
```

___

## 🔒 3. Attributs privés et conventions

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance  # convention d’attribut "protégé"

    def deposit(self, montant):
        self._balance += montant

    def display_balance(self):
        print(f"Solde de {self.owner} : {self._balance} €")

compte = BankAccount("Jean")
compte.deposit(150)
compte.display_balance()
```

___

## 🧬 4. Héritage (optionnel selon le temps)

```python
class Animal:
    def make_sound(self):
        print("Cet animal fait un bruit.")

class Dog(Animal):
    def make_sound(self):
        print("Ouaf Ouaf !")

class Cat(Animal):
    def make_sound(self):
        print("Miaou !")

my_dog = Dog()
my_dog.make_sound()

my_cat = Cat()
my_cat.make_sound()
```
